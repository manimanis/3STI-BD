```python
import pandas as pd
```

```python
df = pd.read_csv('Fifa_world_cup_matches.csv')
df.head()
```

```python
df.info()
```

```python
df.columns = ["equipe1", "equipe2", "buts_eq1", "buts_eq2", "date", "heure", "categorie", 
"c_jaunes_eq1","c_jaunes_eq2", "c_rouges_eq1", "c_rouges_eq2", "fautes_eq2", "fautes_eq1",
"hors_jeu_eq1", "hors_jeu_eq2", "corner_eq1", "corner_eq2", "penalty_eq1", "penalty_eq2"]
df.head()
```

```python
equipes = pd.concat([df.equipe1, df.equipe2], ignore_index=True)
equipes = equipes.unique()
equipes.sort()
equipes
```

```python
categories = df.categorie.unique()
categories = [categorie for categorie in categories if categorie.startswith('Group')]
categories.sort()
categories
```

```python
equipes_dict = []
for categorie in categories:
    for equipe in equipes:
        res = df[((df.equipe1 == equipe) | (df.equipe2 == equipe)) & (df.categorie == categorie)]['categorie']
        if len(res) != 0:
            equipes_dict.append({
                'equipe': equipe,
                'categorie': categorie
            })
equipes_dict
```
