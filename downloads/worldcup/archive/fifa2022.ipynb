```python
import pandas as pd
```

```python
df = pd.read_csv('Fifa_world_cup_matches.csv')
df.head()
```

```python
df.info()
```

```python
df.columns = ["equipe1", "equipe2", "buts_eq1", "buts_eq2", "date", "heure", "categorie", 
"c_jaunes_eq1","c_jaunes_eq2", "c_rouges_eq1", "c_rouges_eq2", "fautes_eq2", "fautes_eq1",
"hors_jeu_eq1", "hors_jeu_eq2", "corner_eq1", "corner_eq2", "penalty_eq1", "penalty_eq2"]
df.head()
```

```python
equipes = pd.concat([df.equipe1, df.equipe2], ignore_index=True)
equipes = equipes.unique()
equipes.sort()
equipes
```

```python
categories = df.categorie.unique()
categories = [categorie for categorie in categories if categorie.startswith('Group')]
categories.sort()
categories
```

```python
equipes_dict = []
num_eq = 0
for categorie in categories:
    df_eq = df[['equipe1', 'equipe2']][df.categorie == categorie]
    eqps = pd.concat([df_eq.equipe1, df_eq.equipe2], ignore_index=True).unique()
    eqps.sort()
    for equipe in eqps:
        num_eq += 1
        equipes_dict.append({
                'num_equipe': num_eq,
                'equipe': equipe,
                'categorie': categorie
            })
equipes_dict[:5]
```

```python
df_equipes = pd.DataFrame(equipes_dict)
df_equipes
```

```python
df_equipes.to_csv("equipes.csv", index=False)
```

```python
cols = ['equipe1','equipe2','buts_eq1','buts_eq2','date','heure','categorie']
matches = df.loc[:][cols]
matches.head()
```

```python
matches['date_match'] = pd.to_datetime(matches['date'] + " " + matches['heure'], format='%d %b %Y %H:%M:%S')
matches.head()
```

```python
cols = ['equipe1','equipe2','buts_eq1','buts_eq2','date_match','categorie']
matches = matches.loc[:][cols]
matches.head()
```

```python
matches.replace({
    'Group A': 'Play off',
    'Group B': 'Play off',
    'Group C': 'Play off',
    'Group D': 'Play off',
    'Group E': 'Play off',
    'Group F': 'Play off',
    'Group G': 'Play off',
    'Group H': 'Play off'
}, inplace=True)
matches.categorie.unique()
```

```python
def fn(row):
    row["num_eq1"] = sum(eq['num_equipe'] for eq in equipes_dict if eq['equipe'] == row['equipe1'])
    row["num_eq2"] = sum(eq['num_equipe'] for eq in equipes_dict if eq['equipe'] == row['equipe2'])
    return row
matches = matches.apply(fn, axis=1)
matches.head()
```

```python
matches['num_match'] = [i+1 for i in range(len(matches))]
cols = ['num_match', 'categorie', 'num_eq1', 'num_eq2', 'date_match']
matches = matches.loc[:][cols]
matches.head()
```

```python
matches.to_csv('matches.csv', index=False)
```

```python
cols_events = [('buts', 'buts_eq1', 'buts_eq2'), ('carte jaune', 'c_jaunes_eq1', 'c_jaunes_eq2'), 
               ('carte rouge', 'c_rouges_eq1', 'c_rouges_eq2'), ('faute', 'fautes_eq2', 'fautes_eq1'), 
               ('hors jeu', 'hors_jeu_eq1', 'hors_jeu_eq2'), ('corner', 'corner_eq1', 'corner_eq2'), 
               ('penaltiy', 'penalty_eq1', 'penalty_eq2')]
events = []
num_event = 0
for i in range(len(matches)):
    for event, col1, col2 in cols_events:
        num_event += 1
        dct = {
            'num_event': num_event,
            'num_match': i + 1,
            'event': event,
            'equipe1': df.iloc[i][col1],
            'equipe2': df.iloc[i][col2]
        }
        events.append(dct)
events_df = pd.DataFrame(events)
events_df.head(10)
```

```python
events_df.to_csv('events.csv', index=False)
```
